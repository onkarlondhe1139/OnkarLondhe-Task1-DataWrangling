Task 1 - Data Immersion & Wrangling

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("swiggy.csv") 
df.head()
```

Basic Dataset Overview

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
# Check Missing Values
df.duplicated().sum()

# Remove duplicates if needed
df = df.drop_duplicates()
```

```python
# Check Unique Categories
df["City"].unique()
df["Area"].nunique()
df["Food type"].head()
```

```python
# Detect Outliers
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    print(col, ":", df[col].describe(), "\n")
```

Data Cleaning & Standardization

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype=="object" else x)
```

```python
df["city"] = df["city"].str.title()
df["restaurant"] = df["restaurant"].str.title()
```

Create Price Category

```python
df["price_category"] = pd.cut(
    df["price"],
    bins=[0,200,400,700,2000],
    labels=["Low","Medium","High","Premium"]
)
```

Extract Main Food Type

```python
df["primary_food"] = df["food_type"].str.split(",").str[0]
```

 Final Quality Check

```python
df.info()
df.describe()
df.head()
```

Save Cleaned Dataset

```python
df.to_csv("cleaned_restaurant_data.csv", index=False)
```
